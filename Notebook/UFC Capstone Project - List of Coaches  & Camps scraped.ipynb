```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
website_url = 'https://en.wikipedia.org/wiki/List_of_professional_MMA_training_camps'
response = requests.get(website_url)

response.status_code
```

### Next look at the content ot the result 
- it is a `Response` datatype
- but it looks like an html document

```python
print(type(response))
response.content
```

#I'll use the BeautifulSoup package to extract data from 'lxml'

```python
soup = BeautifulSoup(response.content, 'lxml')
print(soup.title)
```

```python
print(soup)
```

### Let's get the table that contains the data from the page using beautiful soup

```python
tables = soup.find_all('table', attrs = {'class': 'wikitable'})
tables
```

```python
###check to see how many tables was scraped
- then use `pd.read_html()` get the list of dataframe extracted from the soup tables
- you'll need to convert the tables (still a response object) to a string before pandas can read it
- Let's load the table to a dataframe
```

```python
type(tables)
```

```python
len(tables)
```

```python
result_list = pd.read_html(str(tables[0])) # a list of dataframes
len(result_list)
```

```python
type(result_list)
```

```python
Coaches = result_list[0] 
Coaches
```

```python
Coaches = Coaches.rename(columns = {'Current Fighters (UFC title holders in bold)' : 'Current Fighters'})
Coaches
```

```python
Coaches['split'] = Coaches.apply(' '.join, axis = 1)
Coaches[['Camp(in alphabetic order)', 'Coaches', 'Current Fighters', 'Previous Fighters', 'Location']] = Coaches['Split'].str.extract('([A-Z ]+)\s+([A-Z]{2})\s+(\d+)', expand = True)
Coaches.drop('split', axis = 1, inplace = False)
```

```python
split = lambda x:pd.Series([i for i in reversed(x.split(','))])
```

```python
Location_df = Coaches['Location'].apply(split)
```

```python
Location_df
```

```python
fry = lambda x:pd.Series([i for i in reversed(x.split(','))])
Coach_df = Coaches['Coaches'].apply(fry)
```
