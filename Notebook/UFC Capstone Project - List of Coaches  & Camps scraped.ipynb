```python
import requests
from bs4 import BeautifulSoup 
import re
import pandas as pd
```

```python
website_url = 'https://en.wikipedia.org/wiki/List_of_professional_MMA_training_camps'
response = requests.get(website_url)

response.status_code
```

### Next look at the content ot the result 
- it is a `Response` datatype
- but it looks like an html document

```python
print(type(response))
type(str(response.content).replace('</a><br/>',' '))
```

#I'll use the BeautifulSoup package to extract data from 'lxml'

```python
soup = BeautifulSoup(response.content, 'lxml')
print(soup.title)
```

```python
print(soup)
```

### Let's get the table that contains the data from the page using beautiful soup

```python
tables = soup.find_all('table', attrs = {'class': 'wikitable'})
tables
```

###check to see how many tables was scraped
- then use `pd.read_html()` get the list of dataframe extracted from the soup tables
- you'll need to convert the tables (still a response object) to a string before pandas can read it
- Let's load the table to a dataframe

```python
type(tables)
```

```python
len(tables)
```

```python
result_list = pd.read_html(str(response.content).replace('</a><br/>',' '), attrs = {'class': 'wikitable'}) # a list of dataframes
len(result_list)
```

```python
re.findall('[A-Z][^A-Z]*', 'TheLongAndWindingRoad')
['The', 'Long', 'And', 'Winding', 'Road']
re.findall('[A-Z][^A-Z]*', 'ABC')
['A', 'B', 'C']
```

```python
type(result_list)
```

```python
Coaches = result_list[0] 
Coaches
```

```python
Coaches.rename(columns={'Camp(in alphabetic order)\n': 'Camps', 'Coaches\n':'Coaches', 'Current Fighters\n(UFC title holders in bold)\n\n':'Current Fighters', 'Previous Fighters\n':'Previous Fighters', 'Location\n':'Location'},inplace=True)
Coaches.head()
```

```python
Coaches = Coaches.rename(columns = {'Current Fighters (UFC title holders in bold)' : 'Current Fighters'})
Coaches
```

```python
Coaches.to_csv('Coaches_list.csv', index=False)
```
