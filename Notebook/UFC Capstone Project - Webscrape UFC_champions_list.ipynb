```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
import numpy as np
import re
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import seaborn as sns
import urllib.robotparser as urobot
from urllib.request import urlopen
from pylab import rcParams

%matplotlib inline
```

```python
#Base URL needed for my analysis

url_rank1 = 'https://en.wikipedia.org/wiki/List_of_UFC_champions#Men'

html = urlopen(url_rank1)
```

```python
#Create a BS object from the html by passing html to the BS() fx via the parser 'lxml'

soup = BS(html, 'lxml')
type(soup)
```

```python
#let's scrape the title via the soup object

title = soup.title
print(title)
```

```python
#test to see txt appearing in the webpage and print it out

text = soup.get_text()
print(soup.text)
```

```python
soup.find_all('a')
```

```python
#now use the soup object to scrape the html tags w/in the webpage "a = hyperlinks", "tr/th = table rows/header","td = table cells" 
#use a for loop and get the href method to extract the hyperlinks

all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
#print out the rows only

rows = soup.find_all('tr')
print(rows[:10])
```

```python
#get all table rows & convert to a df w/ a for loop that iterates through table rows

for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
#remove the html tags by passing the cells string into BS w/ get_text() method

str_cells = str(row_td)
cleantext = BS(str_cells, 'lxml').get_text()
print(cleantext)
```

```python
#generate an empty list from the characters specified extract the txt

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>') #compile regular expression by passing a string to match with this argument re.compile()
    clean2 = (re.sub(clean, '',str_cells)) #re.sub method finds all substrings where reg expression matches replace an empty str.
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
#adjust the format by splitting "0" column into multiple ones at the , position

df1 = df[0].str.split(',', expand=True)
df1.head(10)
```

```python
#Add the columns header

col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BS(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head(3)
```

```python
#remove "["

df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```
